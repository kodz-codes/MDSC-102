# **Before Normalising**

```python
# Function for all the plots(Boxplot, Probability, Histogram) before Normalising
def all_plot_before_normalizing(df):
  j = 1
  for i in df.columns:
    plot.figure(figsize = (15,50))
    if i != 'class':
      plot.subplot(13,3,j)
      # title_string = i+"Box Plot"
      sns.boxplot(df, x = df[i]).set(title = ""+i+" Box Plot")
      j += 1
      plot.subplot(13,3,j)
      sns.histplot(data = df, x = df[i], kde = True).set(title = ""+i+" Histogram Plot")
      j += 1
      plot.subplot(13,3,j)
      sp.stats.probplot(df[i], plot = plot)
      j += 1
```

```python
all_plot_before_normalizing(data)
```

# **After Normalising**

```python
def all_plot_after_normalizing(df):
  for i in df.columns:
    j = 1
    Boxcox_lambda = sp.stats.boxcox(df[i])
    lambda_value = Boxcox_lambda[1]
    df[i] = ((df[i] ** lambda_value) - 1)/ lambda_value

    #plotting
    plot.figure(figsize = (15,50))
    if i != 'class':
      plot.subplot(13,3,j)
      # title_string = i+"Box Plot"
      sns.boxplot(df, x = df[i]).set(title = ""+i+" Box Plot")
      j += 1
      plot.subplot(13,3,j)
      sns.histplot(data = df, x = df[i], kde = True).set(title = ""+i+" Histogram Plot")
      j += 1
      plot.subplot(13,3,j)
      sp.stats.probplot(df[i], plot = plot)
      j += 1
```

```python
all_plot_after_normalizing(data)
```

```python
def compare(df):
  df1 = df
  for i in df1.columns:
    j, k = 1,1
    # plotting before normaisation
    plot.figure(figsize = (15,50))
    if i != 'class':
      plot.subplot(13,3,j)
      # title_string = i+"Box Plot"
      sns.boxplot(df, x = df1[i]).set(title = ""+i+" Before Normalizing Box Plot")
      j += 1
      plot.subplot(13,3,j)
      sns.histplot(data = df1, x = df1[i], kde = True).set(title = ""+i+" Before Normalizing Histogram Plot")
      j += 1
      plot.subplot(13,3,j)
      sp.stats.probplot(df1[i], plot = plot)
      j += 1

    Boxcox_lambda = sp.stats.boxcox(df1[i])
    lambda_value = Boxcox_lambda[1]
    df1[i] = ((df1[i] ** lambda_value) - 1)/ lambda_value
    # plotting after normaisation
    plot.figure(figsize = (15,50))
    if i != 'class':
      plot.subplot(13,3,k)
      # title_string = i+"Box Plot"
      sns.boxplot(df1, x = df1[i]).set(title = ""+i+" After Normalizing Box Plot")
      k += 1
      plot.subplot(13,3,k)
      sns.histplot(data = df1, x = df1[i], kde = True).set(title = ""+i+" After Normalizing Histogram Plot")
      k += 1
      plot.subplot(13,3,k)
      sp.stats.probplot(df1[i], plot = plot)
      k += 1
```

```python
compare(data)
```
