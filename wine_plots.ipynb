```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
column_names = ['class','AlcoholMalic', 'acid','Ash','Alcalinity of ash', 'Magnesium', 'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color_intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
#Loading the file into the code
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDSC-102/wine/wine.data',names = column_names)
```

```python
#displaying the first 5 rows of the data
data.head()
data
```

```python
# Count plot for the feature 'Alcalinity of ash
sns.countplot(x = data['Alcalinity of ash'])
plot.show()

# Count plot for the feature 'class' of the alcohols
sns.countplot(data = data, x = 'class')
plot.title("Count plot of the feature 'Class' of alcohol")
```

```python
# Scatter plot between two features - Acid and Alcalinity of ash based on their class
sns.scatterplot(data = data, x = 'acid', y = 'Alcalinity of ash', hue = 'class', palette=['blue','red','black'])
plot.title("Scatter plot between acid and Alcalinity of ash based on their class")
```

Inference for the above scatter plot:

We can clearly see that the first class of alcohols are having less content of acid but a moderate content of alcalinity of ash, wehre as comparitively third class of alcohols have high contents od both acid and alcalinity of ash compared to all other three classes of alcohol


```python
# Plot for protein content in alochol versus all other attributes based on their class using scatter plot.
fig, axes = plot.subplots(4,3, figsize = (30,40)) # subplots no. of rows, columns and size of each subplot
# fig.subplots_adjuts(hspace = 1.0, wspace = 1.0 )# required height and width space between two sub plots
i = 1
for feature in data:
  if feature != 'OD280/OD315 of diluted wines' and feature != 'class':
    plot.subplot(4,3,i)
    sns.scatterplot(x = data[feature], y = data['OD280/OD315 of diluted wines'], hue = data['class'])
    plot.grid()
    i = i+1
```
