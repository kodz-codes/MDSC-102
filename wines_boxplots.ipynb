```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
column_names = ['class','AlcoholMalic', 'acid','Ash','Alcalinity of ash', 'Magnesium', 'Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color_intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
#Loading the file into the code
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDSC-102/wine/wine.data',names = column_names)
```

```python
sns.boxplot(data = data, x = "class")
```

```python
sns.boxplot(data = data, x = "class", y = "acid", hue = "class")
```

```python
sns.boxplot(data = data, y = "AlcoholMalic", x = "class", hue = "class")
plot.grid()
```

```python
fig, axes = plot.subplots(5,3, figsize = (30,40))
i = 1
for feature in data:
  if feature != 'class':
    plot.subplot(5,3,i)
    sns.boxplot(data,x = "class",y = data[feature], hue = "class")
    # plot.grid()
    i = i+1
```
