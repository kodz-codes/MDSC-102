```python
#importing the neccesary libraries

import numpy as np
import scipy as sp
from scipy.stats import binom
import matplotlib.pyplot as plt
```

# Binomal distribution plot

```python
# defining the paramters for the plot
#number of trials
trail_num = int(input("Enter the number of trials to be performed :"))
# Probabilty of success
succ_prob = float(input("Enter the success probability : "))

x_values = np.arange(0,trail_num+1) # Possible values X can assume

Probabilities = np.array([np.math.comb(trail_num, x) * (succ_prob**x) * ((1-succ_prob)**(trail_num-x)) for x in x_values])

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution")
#plt.plot(X_values,probabilities)
plt.bar(x_values,Probabilities)
```

# Exponential Distribution plot

```python
# Define the parameters
# Taking the lamba value
lamda= float(input("Enter the lamba value = ")) 

#possible vaules x can assume
x_values = np.arange(-10,11,.2) 

probabilities = np.array([(lamda * np.exp(-lamda*x)) for x in x_values])

plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Exponential Distribution")
#plt.plot(X_values,probabilities)
plt.plot(x_values,probabilities)
```

# Normal Distribution Plot

```python
# Defining the parameters
u = float(input("Enter the standard deviation = ")) 
mean = float(input("Enter the mean = "))  

pi = 3.1415

#possible vaules x can assume
x_values = np.arange(-10,11,.2)  



probabilities = np.array([1/np.sqrt(2*pi*mean**2) * np.exp(-(x-u)**2/2*mean**2) for x in x_values])


plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")
plt.plot(x_values,probabilities)
#plt.bar(X_values,probabilities)
```

# Gamma Distribution Plot

```python
# Defining the parameters
alpha = float(input("Enter the alpha = ")) 
theta = float(input("Enter the lamda = "))

#possible vaules x can assume
X_values = np.linspace(0, 10, 1000) 

def gamma_pdf(x, alpha, theta):
    numerator = (x ** (alpha - 1)) * np.exp(-x * theta)
    denominator = (theta ** alpha) 
    return numerator / denominator


# Calculate the manual gamma PDF for each x value
probabilities = np.array([gamma_pdf(x,alpha, theta) for x in X_values])


#plt.plot(x, pdf, label=f"Gamma(k={k}, theta={theta})")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Gamma Distribution")
plt.plot(X_values,probabilities)
#plt.bar(X_values,probabilities)
plt.show()
```
